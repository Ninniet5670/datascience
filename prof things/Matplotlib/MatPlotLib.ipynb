## Introduction

Matplotlib is the "grandfather" library of data visualization with Python. It was created by John Hunter. He created it to try to replicate MatLab's (another programming language) plotting capabilities in Python. So if you happen to be familiar with matlab, matplotlib will feel natural to you.

It is an excellent 2D and 3D graphics library for generating scientific figures. 

Some of the major Pros of Matplotlib are:

* Generally easy to get started for simple plots
* Support for custom labels and texts
* Great control of every element in a figure
* High-quality output in many formats
* Very customizable in general

Matplotlib allows you to create reproducible figures programmatically. Let's learn how to use it! Before continuing this lecture, I encourage you just to explore the official Matplotlib web page: http://matplotlib.org/

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.linspace(0,5,11)
```

```python
y = x**2
```

```python
x
```

```python
y
```

```python
plt.plot(x,y,"r-")
```

```python
plt.plot(x,y,"r-")
plt.xlabel("X LABEL")
plt.ylabel("Y LABEL")
plt.title("TITLE")
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
#OBJECT ORIENTED

fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel("X LABEL")
axes.set_ylabel("Y LABEL")
axes.set_title("OBJECT ORIENTED GRAPH")
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes1.plot(x,y)
axes1.set_title("LARGER PLOT")

axes2 = fig.add_axes([0.2,0.5,0.4,0.3])
axes2.plot(y,x)
axes2.set_title("SMALLER PLOT")
```

```python
#SUBPLOTS

fig,axes = plt.subplots(nrows=1, ncols=2)

axes[0].plot(x,y)
axes[0].set_title("AXES1")

axes[1].plot(y,x)
axes[1].set_title("AXES2")
```

# Figure Size & DPI

```python
fig = plt.figure(figsize=(8,2))

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=1, ncols=2, figsize=(4,2))

axes[0].plot(x,y)
axes[1].plot(y,x)

plt.tight_layout()
```

# Saving A Figure

```python
fig = plt.figure(figsize=(4,2))

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig.savefig("mypicture.jpg", dpi=200)
```

# Adding A Legend To Your Plot

https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.legend.html

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x**2,label="X Squared")
ax.plot(x,x**3,label="X Cubed")

ax.legend(loc=(0.1,0.1))
```

# Plot Customization

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color="#FF8C00", lw=2, ls="-.", marker="o", markersize=25, markerfacecolor="blue",
       markeredgewidth=5, markeredgecolor="red") #HEX CODE
```

# Control Over Axis Appearance / Plot Range

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color="blue", lw=2, ls="-.") 

ax.set_xlim([0,1])
ax.set_ylim([0,2])
```
